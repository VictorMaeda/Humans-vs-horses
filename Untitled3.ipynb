```python
BATCH_SIZE = 64
IMG_SIZE = (224, 224)
PATH_DIR = path + "/horse-or-humans-dataset"
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import Model
from tensorflow.keras.layers import Dense
from tensorflow.keras.applications import VGG16
import matplotlib.pyplot as plt
```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.3,
    horizontal_flip=True,
)

train_generator = train_datagen.flow_from_directory(
    "/root/.cache/kagglehub/datasets/sanikamal/horses-or-humans-dataset/versions/1/horse-or-human/train",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="binary",
)

valid_datagen = ImageDataGenerator(
    rescale=1.0/255
)
valid_generator = valid_datagen.flow_from_directory(
    '/root/.cache/kagglehub/datasets/sanikamal/horses-or-humans-dataset/versions/1/horse-or-human/horse-or-human/validation',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'
)
```

```python
vgg = keras.applications.VGG16(weights='imagenet', include_top=True)
vgg.summary()
```

```python
inp = vgg.input

new_classification_layer = Dense(1, activation='sigmoid')

out = new_classification_layer(vgg.layers[-2].output)

model_new = Model(inp, out)
```

```python
# make all layers untrainable by freezing weights (except for last layer)
for l, layer in enumerate(model_new.layers[:-1]):
    layer.trainable = False

# ensure the last layer is trainable/not frozen
for l, layer in enumerate(model_new.layers[-1:]):
    layer.trainable = True

model_new.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model_new.summary()
```

```python
history = model_new.fit(
    train_generator,
    epochs=10,
    validation_data=valid_generator
   # ,callbacks=[myCallback, reduce_lr]
)
```

```python
fig = plt.figure(figsize=(16,4))
ax = fig.add_subplot(121)
ax.plot(history.history["val_loss"])
ax.set_title("validation loss")
ax.set_xlabel("epochs")

ax2 = fig.add_subplot(122)
ax2.plot(history.history["val_accuracy"])
ax2.set_title("validation accuracy")
ax2.set_xlabel("epochs")
ax2.set_ylim(0, 1)

plt.show()
```
